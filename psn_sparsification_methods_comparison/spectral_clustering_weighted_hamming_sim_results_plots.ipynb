```python
import numpy as np
import matplotlib.pyplot as plt

def compute_k_neighbors_range(sample_size):
    scale = sample_size / 10000
    k_neighbors_range = (scale * np.array([300, 500, 600, 800, 1000, 2000, 3000, 5000])).astype(int)
    return k_neighbors_range

def plot_ri_vs_k(ax, k_neighbors_range, ri_sample, sample_size, c):
    ax.plot(k_neighbors_range, ri_sample, label=f'Sample {sample_size//1000}k', marker='o', color=c)
    ax.set_xlabel('k')
    ax.set_ylabel('Rand Index (RI)')
    # ax.set_title(f'Rand Index vs k Neighbours for {sample_size//1000}k Sample Size')
    ax.set_title(f'Sample size: {sample_size//1000},000')
    ax.set_ylim(0.49, 0.75)
    # ax.legend()
    ax.grid(True)

# Data
ri_samples = {
    6000: [0.5143, 0.5009, 0.5009, 0.7184, 0.7184, 0.5948, 0.7158, 0.7114],
    10000: [0.5145, 0.7112, 0.7112, 0.7112, 0.7112, 0.5946, 0.7097, 0.7066],
    20000: [0.5110, 0.7242, 0.5012, 0.7242, 0.7242, 0.5939, 0.7251, 0.7225]
}

# Sample sizes
sample_sizes = [6000, 10000, 20000]

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
colours = plt.rcParams['axes.prop_cycle'].by_key()['color']
i = 0
for ax, sample_size in zip(axs, sample_sizes):
    k_neighbors_range = compute_k_neighbors_range(sample_size)
    ri_sample = ri_samples[sample_size]
    plot_ri_vs_k(ax, k_neighbors_range, ri_sample, sample_size, colours[i])
    i += 1

fig.suptitle('Rand Index vs k neighbours for different sample sizes')
# Adjust layout
plt.tight_layout()
plt.savefig('figures/ri_vs_k.pdf')

# Show plot
plt.show()
```
