```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
importance_general = pd.read_pickle('rf_asd_subgroups/sorted_feature_importances_RF_mf_newest.pkl')
importance_all_data = pd.read_pickle('rf_asd_subgroups/sorted_feature_importances_all_data_multiclass.pkl')
importance_asd_only = pd.read_pickle('rf_asd_subgroups/sorted_feature_importances_asd_only.pkl')

# Extract feature names and importances
features_general = importance_general['feature']
importances_general = importance_general['importance']

# make sure the features are in the same order
importance_all_data = importance_all_data.set_index('feature').reindex(features_general).reset_index()
importance_asd_only = importance_asd_only.set_index('feature').reindex(features_general).reset_index()

features_all_data = importance_all_data['feature']
importances_all_data = importance_all_data['importance']

features_asd_only = importance_asd_only['feature']
importances_asd_only = importance_asd_only['importance']

# Reverse the order of features and importances
features_general = features_general[::-1]
importances_general = importances_general[::-1]

features_all_data = features_all_data[::-1]
importances_all_data = importances_all_data[::-1]

features_asd_only = features_asd_only[::-1]
importances_asd_only = importances_asd_only[::-1]

# Create an index for the y positions
y_pos = np.arange(len(features_general))

# Plotting
fig, ax = plt.subplots(figsize=(8, 7))
h = 0.25
# General importance
ax.barh(y_pos + h*2, importances_general, color='#FFB6C1', label='ASD vs. non-ASD', height=h)

# All data importance
ax.barh(y_pos + h, importances_all_data, color='purple', label='Multiclass for all individuals\' clusters', height=h, alpha=0.7)

# ASD only importance
# ax.barh(y_pos, importances_asd_only, color='green', label='Multiclass for ASD clusters', height=h, alpha=0.5)
ax.barh(y_pos, importances_asd_only, color='orange', label='Multiclass for ASD clusters', height=h, alpha=0.5)

# Adding labels and title
ax.set_yticks(y_pos + h)
ax.set_yticklabels(features_general)
ax.set_title('Feature Importances for different RF Classifiers')
ax.set_xlabel('Mean decrease in impurity')
ax.set_ylabel('Feature')
ax.grid(True, axis='x')
ax.legend(title='Model', loc='lower right')

# Adjust layout
plt.tight_layout()

# plt.savefig('figures/feature_importances_comparison_green.pdf')
plt.savefig('figures/feature_importances_comparison_orange.pdf')

# Show plot
plt.show()
```
